# Лекция №1. Введение в искуственный интеллект.

Информационная революция - технологическая революция, которая до сих пор продолжает развиваться и влияет непосредственно на нашу жизнь. Эта революция фундаментально изменила нашу жизнь, переложив большую часть рутинной работы на компьютеры.

Однако, долгое время не любую задачу можно было поручить компьютеру, поскольку многие задачи подразумевали не только участие человека, но и процесс его мышления, анализа, именно для решения этой проблемы возникла дисциплина искуственного интеллекта (ИИ).

ИИ - научная дисциплина, которая должна научить компьютер решать человекоподобные задачи. Изначально, не было понятно, как обучить компьютер играть в шахматы, шашки и т.д. Этот этап был пройден и ИИ перешло к изучению новых задач, можно сказать, что ИИ занимается решением задач, которые превращают компьютер в человека.

Искуственный интеллект, занимающийся конкретными задачами (преобразование текста в речь, классификация предметов на фото и тд), называют слабым (прикладным) искуственным интеллектом, то есть направленным на решение конкретной, заранее известной задачи. 

Сильный искуственный интеллект (General AI) - ИИ, который может имитировать мышление человека во многих сферах одновременно, он не затачивается под конкретную задачу, сфера его функционала обширна(ChatGPT, Нейро). На данный момент - это очень популярная тема, которую все обсуждают, а также обсуждается возможность и последствия достижения сверхчеловевеческих навыков.

В классическом программирования, ориентированном на вычислительные задачи или задачи, связанные с хранением, обработкой передачей информации (интернет), используются алгоритмы (определённый набор правил), для выполнения определенной задачи. В задачах ИИ, например классификация человека по фото, заранее известного алгоритма не существует (пр. как преобразовать набор пикселей в возраст числа). Но, несмотря на это, человек умеет определять с определённой точностью возраст человека или врач умеет определять диагноз, делается это исходя из набора признаков, опыта, интуиции. 

Искуственный интеллект занимается составлением алгоритмов для подобного рода задач, более формально: 
Это раздел компьютерных наук, занимающихся воспроизведением в ПК рассуждений и способов решения задач, характерных для человека.

Важно:
- ИИ - наука (не термин из научпопа)
- Круг задач, решаемых ИИ постоянно меняется (прогрессирует от более примитивных к усложённым мыслительным процессам)

Задачи решаемые ИИ:
- Распознавание рукописного шрифта (цифр, букв)
- Распознавание шрифта Брейля (для слепых и слабовидящих)
- Определение размеров объектов на фото
- Генерация изображений, текста (Шедеврум, GPT)
- Задачи коллективного поведения (простое поведение каждого объекта в группе ведёт к сложному поведению всей группы).

Популярность и развитие ИИ в последнее время связано с тем, что в последние 10 лет подход к ИИ изменился. Изначально, мы пытались научить ИИ мыслить, как человек, теперь же, активно развивается машинное обучение, в котором искуственному интеллекту дают необходимые данные, результат, а компьютер сам в них разбирается и находит между ними логические закономерности.
За последние годы данных стало в разы больше, большая часть информации оцифровывается, кол-во этой самой информации увеличивается каждый год почти вдвое. Помимо этого, увеличились и вычислительные мощности компьютеров, количество нейроново мозга, которые может эмулировать ПК почти приблизилось к кол-ву нейронов человеческого мозга. При чём, рост вычислительных мощностей происходит экспоненциально. Это всё привело к развитию машинного обучения и ИИ в частности.

## ImageNet Large Scale Visual Recognition Challenge (ILSVRC)

Для понимания того, как развивается ИИ, можно взглянуть на популярную задачу ILSVRC, суть которой заключается в классификации 100 млн картинок на 1000 классов.
Среди классов можно встретить не только общие понятия (кошка, собака), а даже их конкретные породы, вариации, что в разы усложняет задачу классификации. По этому соревнованию, погрешность определения класса в 2010 году была на уровне 30%, а в 2015 году погрешность приравнялась к уровню человека.

P.S. Top-5 Accuracy - это правило погрешности определения класса компьютером. То есть, для задачи классификации ПК выдаёт вероятность принадлежности объекта каждому из возможных классов, если правильный ответ входит в топ-5 этих классов, то классификация считается успешной.

P.S.S. Приравнивание значений погрешности к уровню человека не означает, что ИИ видит лучше, чем человек, тут работает человеческий фактор, человек не знает каких-то пород кошек, ему лень, он устал после 100 картинок и т.д.

 За последние годы ИИ научился:
 - Распознавать речь
 - Синтез речи
 - Машинный перевод
 - Понимание текста (ответы на заданные вопросы)
 - Творческие задачи (лучше 90% людей создаёт изображения по описанию)

## Интеллект. Свойственен ли интеллект только человеку?

Это философский вопрос, поскольку у интеллекта нет научного определения. Тогда, как определить, создали мы искуственный интеллект или нет?

Для этого был создан тест Тьюринга, когда интеллектуальность человека измеряется другим человеком. Берутся люди - испытатели системы, если они не могут сказать, кто именно с ними общается (ИИ или человек), то считается, что это искуственный интеллект.

Особенности теста Тьюринга:
- Позволяет свести философский вопрос к понятной процедуре общения (+)
- Появляется необходимость подстраивать ИИ под человека, что возможно является неправильным, даже ненужным (-)
- Тестируется исключительно способность текстового общения (-)
- Зависит от наблюдателя (-). Исправляется путём изменения выборки наблюдателей, дополнения её большим кол-вом умных людей.

В 2014 году был создан ИИ "Eugene Goostman", которая прошла тест Тьюринга (однако, она манипулировала сознанием людей, выдавая себя за умственного отсталого человека). То есть, система был не столько умной, сколько достаточно хитрой, чтобы обмануть человека. Это и есть ещё один минус теста Тьюринга.

## Языковые модели и GPT-3.

Языковые модели - модели, которые берут начало текста и пытаются его продолжить так, чтобы текст был похож на то, что пишет обычно человек.
Но, при этом, подобные модели не ориентируются на то, что они когда-то либо видели, знали и т.д. То есть, языковая модель выдаёт правдоподобный текст, но необязательно ПРАВДИВЫЙ.

С использованием таких моделей были написаны даже книги.

## Возможно ли создание ИИ?

Гипотеза Ады Лавлейс: "Компьютер программируется человеком, он не способен на то, что не заложено программой"
Однако, гипотеза Ады частично уже опровергнута машинным обучением, которое ведёт себя непредсказуемо от большого кол-ва данных (работает местами так, как её не обучали изначально).

Гипотеза Ньюэла-Саймона: "Для достижения человеческого интеллекта необходимо и достаточно обладать способностью к символьным рассуждениям"

Опровергнуть или подтвердить эту гипотезу ещё не удалось. Однако, у человека наблюдается куда больше способностей, чем просто символьные рассуждения.

Джон Сёрэл предложил задачу "Китайская комната", суть которой в том, что можно выдавать осознанный ответ, не понимая текста. Такой задачей Джон показывал невозможность существования ИИ, однако уже сейчас было доказано обратное. ChatGPT выдаёт осозанные ответы, не понимая текста. Однако вопрос понимания текста нами и нейросетью является уже философским и строго определить его на данный момент не представляется возможным.

## Подходы к созданию ИИ

Нисходящий подход - моделировние в ПК человеческий знаний, рассуждений (то с чего начиналось развитие ИИ). Работает с помощью представления знаний и поиска среди них.

Восходящий (нейрокибернетический) подход - моделирование низкоуровневых процессов в человеческом мозге. Работает, как нейронная сеть.

У обоих подходов имеются минусы. Для нисходящего минусом является - сложность объяснения собственных знаний человеком (многие знания абстрактны и их тяжело описать). У нейросетей (восходящего) - сложность заключалась в вычислениях, то есть вычислительных мощностей не хватало. Из-за развития железа, данный подход становится вновь актуальным.

Существует универсальный подход, который совмещает оба подхода выше - машинное обучение. 
Машинное обучение - обучение ПК решению задач на основе данных или экспериментов. Оно работает с помощью извлечения знаний из данных, построения моделей.


Дополнительные подходы:
- Многоагентный подход - решение сложной задачи множеством простых агентов за счёт синергетического эффекта
- Эволюционный подход - генетические алгоритмы, моделирование эволюции.
- Гибридный подход - совмещение нисходящего и восходящего подхода.

## Сравнение символьного и нейросетевого подхода:

Символьный:
- Нужно извлечение данных из эксперта
- Интерпретируемая модель, то есть можно объяснить почему были приняты решения
- Хорошо описывает процессы, где люди явно рассуждают (принятие решений, планирование)

Нейросетевой:
- Нужно много данных для обучения
- Неинтерпретируемая модель
- Хорошо описывает процессы, происходящие подсознательно (распознание предметов, речь)
